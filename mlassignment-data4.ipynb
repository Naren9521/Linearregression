```python
#Modules used 
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("/kaggle/input/linearregression/Data4.csv")
df = df.drop(columns='Sl.')
```

```python
y = df['y']
y_array = np.array(y)
```

```python
df = df.drop(columns = 'y')
n = 3
```

```python
A = np.zeros((n,n))
for i in range(len(df)):
    temp = np.array(df.iloc[i])
    x1 =temp.reshape((n,1))
    x2 =x1.T 
    A += x1@x2
A
```

```python
A_inverse = np.linalg.pinv(A)
A_inverse
```

```python
sigma_yi_xi = np.zeros((n,1))
for i in range(len(df)):
    temp = np.array(df.iloc[i])
    x1 = temp.reshape((n,1))
    x2 = y_array[i].reshape((1,1))
    sigma_yi_xi+=x1@x2
sigma_yi_xi
```

```python
w = A_inverse@sigma_yi_xi
w
```

```python
constant = 1.5  
lst = []
for i in range(len(df)):
    lst.append(w.T @ np.array(df.iloc[i][:]) + constant)
print(lst)
```

```python
df['y_hat'] = lst
df['y'] = y_array
```

```python
def RSS(df):
    return np.sum((df['y']-df['y_hat'])**2)

def TSS(df):
    return np.sum((df['y']-np.mean(df['y_hat']))**2)
def Rsquare(df):
    return 1 - RSS(df) / TSS(df)
print("Rss",RSS(df))
print("TSS",TSS(df))
```

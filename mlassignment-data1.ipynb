```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv("/kaggle/input/linearregression/Data1.csv")
df
```

```python
y = df['y']
y_array = np.array(y)
df = df.drop(columns='y')
```

```python
df
```

```python
A = np.zeros((1,1))
for i in range(len(df)):
    temp = np.array(df.iloc[i])
    x1 = temp.reshape((1,1))
    x2 = x1.T
    A+= x1@x2
A
```

```python
A_inverse = np.linalg.inv(A)
A_inverse 
```

```python
sigma_yi_xi = np.zeros((1,1))

for i in range(len(df)):
    temp = np.array(df.iloc[i])
    x1 = temp.reshape((1,1))
    x2 = y_array[i].reshape((1,1))
    sigma_yi_xi+= x1*x2
print(sigma_yi_xi)
```

```python
w = A_inverse@sigma_yi_xi
w
```

```python
c = 0.5
lst = []
for i in range(len(df)):
    lst.append(w.T@np.array(df.iloc[i][:] + c))
lst
```

```python
df['y_hat'] = lst
df['y'] = y_array
```

```python
def R_square(df):
    return 1 - RSS(df)/TSS(df)
def RSS(df):
    return np.sum((df['y']-df['y_hat'])**2)
def TSS(df):
    return np.sum((df['y']-np.mean(df['y_hat']))**2)
print("Rss",RSS(df))
print("TSS",TSS(df))
```

```python
df = df.drop(columns ='y')
```

```python
def gradient(w):
    temp = 0
    sdf = df.sample(frac = 1,axis =1 ,random_state=42)
    for i in range(len(sdf)):
        r = np.array(sdf.iloc[i][:])
        yhat = r@w
        temp+= -2*(y[i]-yhat)*sdf.iloc[i][0]/len(sdf)
    return temp

lr = 0.00001
wi = np.zeros((1,1))
while True:
    wn = wi - lr*gradient(wi)
    up = wi - wn
    down = wn
    if np.linalg.norm(up)/np.linalg.norm(down) <= 1e-4:
        break
        
    wi = wn
w = wi
print(w)
    
```

```python
c = 4 
def predict(df,w):
    lst = []
    for i in range(len(df)):
        res = np.array(df.iloc[i][:])
        res2 = res.reshape((1,-1))
        prediction = w.T @ res2 + c
        lst.append(prediction.flatten())
    return np.array(lst)
result = predict(df,w)
```

```python
df['predicted'] = predict(df,w)
df['y'] = y
```

```python
def R_square(df):
    return 1 - RSS(df)/TSS(df)
def RSS(df):
    return np.sum((df['y'] - df['predicted'])**2)
def TSS(df):
    return np.sum((df['y']-np.mean(df['predicted']))**2)

R_square(df)
print("Rss",RSS(df))
print("TSS",TSS(df))
```
